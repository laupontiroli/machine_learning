**Exercise** Exploring linear regression

a) Make an artificial test case: write a function that receives an integer $n$ and

- Generates $n$ random values uniformly distributed in the interval $[0.0, 2.0]$
- Computes $y = 4 + 3x$
- Adds a Gaussian noise to $y$, of zero mean and unit standard deviation
- Returns $x$ and $y$

```python
import numpy as np

def make_test_data(n):
    x = np.random.uniform(0.0, 2.0, [n, 1])
    noise = np.random.normal(0.0, 1.0, [n, 1])
    y = 4.0 + 3.0 * x + noise
    return x, y
```

```python
x, y = make_test_data(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 2)
plt.show()
```

b) Generate $n=10$ points and plot the points. Fit a linear regression model to this data. Record the fit parameters. Now repeat the procedure a total of $1000$ times and plot the histogram of the intercept and the coefficient.

```python
from sklearn.linear_model import LinearRegression

n = 10

theta_0_list = []
theta_1_list = []

for _ in range(1000):
    x, y = make_test_data(n)
    model = LinearRegression()
    model.fit(x, y)
    intercept, slope = model.intercept_[0], model.coef_[0][0]
    theta_0_list.append(intercept)
    theta_1_list.append(slope)

```

```python
plt.hist(theta_1_list, bins=30, range=(0, 8));
```

```python
plt.scatter(theta_0_list, theta_1_list, alpha=0.5)
plt.plot([4.0], [3.0], marker='x', markersize=10, color='red')
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.show()
```

**Exercise** Exploring polynomial regression

a) Make an artificial test case: write a function that receives an integer $n$ and

- Generates $n$ random values uniformly distributed in the interval $[-3.0, 3.0]$
- Computes $y = 0.5 x^2 + x + 2$
- Adds a Gaussian noise to $y$, of zero mean and unit standard deviation

```python
import numpy as np

def make_test_data(n):
    x = np.random.uniform(-3.0, 3.0, [n, 1])
    noise = np.random.normal(0.0, 1.0, [n, 1])
    y = 0.5 * x**2 + x + 2 + noise
    return x, y
```

```python
x, y = make_test_data(1000)
```

```python
import matplotlib.pyplot as plt

x_aux = np.arange(-3, 3, 0.01)
y_aux = 0.5 * x_aux**2 + x_aux + 2

plt.figure(figsize=(4, 3))
plt.scatter(x, y)
plt.plot(x_aux, y_aux, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-3, 3)
plt.show()
```

b) Generate $n=10$ points and plot the points. Fit a linear regression model to this data. Record the fit parameters. Now repeat the procedure a total of $1000$ times and plot the histogram of the parameters.

c) Repeat item (b) but for $n=1000$ points. What do you observe?

```python
from sklearn.linear_model import LinearRegression

n = 1000

def linear_regression(x, y):
    model = LinearRegression()
    model.fit(np.c_[x, x**2], y)
    return model.intercept_[0], model.coef_[0][0], model.coef_[0][1]

theta_0_list = []
theta_1_list = []
theta_2_list = []

for _ in range(1000):
    x, y = make_test_data(n)
    theta_0, theta_1, theta_2 = linear_regression(x, y)
    theta_0_list.append(theta_0)
    theta_1_list.append(theta_1)
    theta_2_list.append(theta_2)
    
```

```python
plt.scatter(theta_0_list, theta_1_list, alpha=0.5)
plt.plot([4.0], [3.0], marker='x', markersize=10, color='red')
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.show()
```
